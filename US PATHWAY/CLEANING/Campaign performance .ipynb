```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(df.isnull().sum())
```

```python
df.drop(['campaign_start_date'], axis=1, inplace=True)
```

```python
df.fillna(df['click_through_rate'].median(),inplace=True)#Using median to fill NA values
```

```python
df.fillna({"adset_name":"Unknown"},inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df.to_excel("D:/LPU/Internship/real data project/updated_camp.xlsx", index=False)#Extracting the cleaned data 
```

```python
# Summary statistics for total_spent
print("Summary statistics for total_spent:")
print(df['total_spent'].describe())

# Histogram of total_spent
plt.figure(figsize=(10, 6))
df['total_spent'].hist(bins=20)
plt.title('Distribution of Total Spent')
plt.xlabel('Total Spent')
plt.ylabel('Frequency')
plt.show()
```
