```python
#Filled with the Nearby values for campaign Name
df['Campaign Name'] = df['Campaign Name'].fillna(method='ffill').fillna(method='bfill')
```

```python
df.isnull().sum()
```

```python
#Mode for graduation degree #Easy For Categorical
mode_graduation_degree = df['graduation_degree'].mode()[0]
df['graduation_degree'].fillna(mode_graduation_degree, inplace=True)

mean_value = df['graduation_degree'].mode()
print(f"The mean value of 'graduation_percentage' is: {mean_value}")
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(6,4))
sns.distplot(df['graduation_percentage'].dropna(), kde=True, bins=30, color='blue')
plt.title('Distribution of Graduation Percentage with KDE', fontsize=16)
plt.xlabel('Graduation Percentage', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
#Mean for graduationpercentage #Easy To Find Avg of it
df['graduation_percentage'].fillna(df['graduation_percentage'].mean(), inplace=True)

mean_value = df['graduation_percentage'].mean()
print(f"The mean value of 'graduation_percentage' is: {mean_value}")
```

```python
plt.figure(figsize=(6,4))
sns.distplot(df['graduation_percentage'].dropna(), kde=True, bins=30, color='blue')
plt.title('Distribution of Graduation Percentage with KDE', fontsize=16)
plt.xlabel('Graduation Percentage', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
#Mode for work_experience # Easy For CategoricaL data
mode_work_experience = df['work_experience'].mode()[0]
df['work_experience'].fillna(mode_work_experience, inplace=True)

mode_value = df['work_experience'].mode()
print(f"The mean value of 'work_experience' is: {mode_value}")
```

```python
df.isnull().sum()
```

```python
#Checking for outlier in the Graduation Percentage column
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(df['graduation_percentage'])
plt.title('Box Plot of Graduation Percentage')
plt.xlabel('Graduation Percentage')
plt.show()
```

```python
#Replaced 
df['lsq_lead_stage'] = df['lsq_lead_stage'].replace('Invalid', 'Invalid phone number')
df['lsq_lead_stage'] = df['lsq_lead_stage'].replace(['Connected/ Not Interested', 'Not Intrested'], 'Cold')

```

```python
df.head()
```

```python
#updates the 'contacted' column based on values in the 'lsq_lead_stage' column, marking entries as 'not contacted' if they are 'invalid number', 'invalid', or 'Wrong Number', and 'contacted' otherwise
df['contacted'] = df['lsq_lead_stage'].apply(
    lambda x: 'not contacted' if x in ['Invalid phone number'] else 'contacted'
)
```

```python
# Fill null values with 'Unknown' in these columns 
df['lsq_lead_stage'] = df['lsq_lead_stage'].fillna('Unknown')
df['lsq_lead_owner']=df['lsq_lead_owner'].fillna('Unknown')
df['lsq_source']=df['lsq_source'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df['mapped']=df['mapped'].fillna('UnMapped')
```

```python
df.isnull().sum()
```

```python
df=df.drop('Ad Name',axis=1)
```

```python
df.isnull().sum()
```

```python
df['form_name']=df['form_name'].fillna('Others')
```

```python
df.isnull().sum()
```

```python
df=df.drop('lead_id.1',axis=1)
```

# BIVARIATE 

```python
# Choose two categorical variables to analyze, for example, 'platform' and 'lsq_source'
categorical_variable1 = 'lsq_lead_owner'
categorical_variable2 = 'lsq_lead_stage'

# Create a contingency table
contingency_table = pd.crosstab(df[categorical_variable1], df[categorical_variable2])

# Display the contingency table
print(contingency_table)

# 1. Heatmap of the Contingency Table
plt.figure(figsize=(25, 8))
sns.heatmap(contingency_table, annot=True, fmt="d", cmap="YlGnBu")
plt.title(f'Heatmap of {categorical_variable1} vs {categorical_variable2}')
plt.xlabel(categorical_variable2)
plt.ylabel(categorical_variable1)
plt.show()
```

```python
contingency_table.plot(kind='bar', stacked=True, figsize=(17, 10))
plt.title(f'Stacked Bar Plot of {categorical_variable1} vs {categorical_variable2}')
plt.xlabel(categorical_variable1)
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title=categorical_variable2)
plt.show()
```

```python
numerical_variable = 'graduation_percentage'
categorical_variable = 'contacted'

plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x=categorical_variable, y=numerical_variable,hue=categorical_variable,legend=False, palette='Set3')
plt.title(f'Violin Plot of {numerical_variable} by {categorical_variable}')
plt.xlabel('Contact Info')
plt.ylabel('Graduation Percentage')
plt.xticks(rotation=360)
plt.show()
```
